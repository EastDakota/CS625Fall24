```python
print(orig_data.columns.values)
dropped_column = orig_data
print(dropped_column['RATING'])
dropped_column['RATING'] = pd.to_numeric(dropped_column['RATING'], errors='coerce')
dropped_column = dropped_column[(dropped_column.RATING <= 10)]
dropped_column = dropped_column[(dropped_column.RATING >= 0)]
print(dropped_column)
```

```python
dropped_column = dropped_column[dropped_column != None]
dropped_column['STARS'] = dropped_column['STARS'].str.replace('\n', '')
dropped_column['STARS'] = dropped_column['STARS'].str.replace('\r', '')
print(dropped_column)
```

```python
dropped_column['YEAR'] = dropped_column['YEAR'].str.replace(r"\s*\([^()]*\)", "", regex=True)
print(dropped_column['YEAR'])
```
